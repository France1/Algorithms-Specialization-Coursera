### Assignment data

```python
coords = [(0,0),(1,0),(1,1),(0,1)]

coords = []
with open('nn.txt') as f:
    for i,l in enumerate(f):
        l = tuple(map(float, l.split()))
        if i!=0:
            coords.append(l[1:])
x,y = zip(*coords)
```

```python
l = 0
start = coords.pop(0)
visited = [start]
s = start
```

```python
while coords:
#     print(s, coords)
    d2 = [(i, (c[0]-s[0])**2+(c[1]-s[1])**2) for i,c in enumerate(coords)]
    d2 = sorted(d2, key = lambda x: (x[1],x[0]))
#     print(d2)
    i_min, d2_min = d2[0]
    visited.append(coords[i_min])
    l += sqrt(d2_min)
    s = coords.pop(i_min)
#     print('')

l += sqrt((s[0]-start[0])**2+(s[1]-start[1])**2)
```
