### PA#1 and PA#2 Huffman code

```python
import heapq 
from anytree import Node, RenderTree
```

```python
# # small test cases
# d = {'A':3, 'B':2, 'C':6, 'D':8, 'E':2, 'F':6}

# h = [(v,k) for k,v in d.items()]
# heapq.heapify(h)

# Nodes = {k:Node(k) for k in d.keys()}
# Nodes

# counts = {k:0 for k in d.keys()}
```

```python
# assignment data
Nodes = {}
h = []
with open('huffman.txt') as f:
    for i,l in enumerate(f):
        if i!=0:
            Nodes[str(i)] = Node(str(i))
            h.append((int(l), str(i)))
            
# initialize heap of weights and characters
heapq.heapify(h)
```

```python
while(len(h)>1):

    w1, c1 = heapq.heappop(h)
    w2, c2 = heapq.heappop(h)
    # merge characters
    c_m = c1+'_'+c2
    # sum weights
    w_m = w1+w2
    # define merged node as parent of c1 & c2
    Nodes[c_m] = Node(c_m)
    Nodes[c1].parent = Nodes[c_m]
    Nodes[c2].parent = Nodes[c_m]
    # push merged character & its weight to heap
    heapq.heappush(h,(w_m,c_m))
```

```python
# visualise graph (small cases only)
root = Nodes[h[0][1]]
for pre, fill, node in RenderTree(root):
    print("%s%s" % (pre, node.name))
```

```python
# collect depths of leaf nodes
depths = [(k,v.depth) for k,v in Nodes.items() if Nodes[k].is_leaf==True]
print('Min depth = ', min(depths, key = lambda i:i[1]))
print('Max depth = ', max(depths, key = lambda i:i[1]))
```

### PA#3 

```python
# test case
weights = [280,618,762,908,409,34,312,277,246,779]
```

```python
# assignment data
weights = []
with open('mwis.txt') as f:
    for i,l in enumerate(f):
        if i!=0:
            weights.append(int(l))
```

```python
# initialize weights
weights = [0,]+weights
n = len(weights)
# 1. Compute max weights
A = [0,]*(n)
A[1] = weights[1]
for i in range(2,n):
    A[i] = max(A[i-1], A[i-2]+weights[i])
# 2. Reconstruct path
S = []
i = n-1
while i >= 1:
    if A[i-1] >= A[i-2]+weights[i]:
        i = i-1
    else:
        S.append(i)
        i = i-2
        
print(sorted(S))
```

```python
# answer to solution
values = [1, 2, 3, 4, 17, 117, 517, 997]
bits = ''

for v in values:
    bits += str(int(v in S))
print(bits)
```
